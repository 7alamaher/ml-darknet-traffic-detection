## Phase 1 — Dataset Acquisition

```python
import pandas as pd

df = pd.read_csv("Darknet 1.csv")

# Remove empty columns created by trailing commas
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

print("Dataset shape:", df.shape)
print("Number of columns:", len(df.columns))
```

## Phase 2 — Data Preprocessing

```python
# Replace infinite values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Remove rows with missing values
df.dropna(inplace=True)

print('Dataset shape after cleaning:', df.shape)
```

```python
# Remove socket related columns
columns_to_remove = [
    'Flow ID',
    'Src IP',
    'Dst IP',
    'Src Port',
    'Dst Port'
]

df.drop(columns=columns_to_remove, inplace=True, errors='ignore')

print('Dataset shape after removing socket columns:', df.shape)
```

## Prepare Labels for Later Phases

```python
# Binary classification labels (Benign vs Darknet)
X_binary = df.drop(columns=['Label', 'Label.1'])
y_binary = df['Label']

# Multiclass classification labels (Tor / NonTor / VPN / NonVPN)
X_multi = df.drop(columns=['Label', 'Label.1'])
y_multi = df['Label.1']
```

```python
# Verify label distribution
print('Binary label distribution:')
print(y_binary.value_counts())

print('\nMulticlass label distribution:')
print(y_multi.value_counts())
```
